# 01 - Criação e Manipulação de Arrays

Crie um array Numpy com 10 números inteiros aleatórios entre 1 e 20 e extraia os elementos pares.

```python
import numpy as np

lista = np.random.randint(1, 21, size=10)

pares = lista[lista % 2 == 0]

print(f"Lista: {lista}")
print(f"Pares: {pares}")
```

# 02 - Operações Aritméticas com Arrays

Crie dois arrays de 10 números inteiros aleatórios entre 0 e 50 e realize operações de soma, subtração, multiplicação e divisão elemento a elemento.

```python
import numpy as np

lista1 = np.random.randint(1, 51, size=10)
lista2 = np.random.randint(1, 51, size=10)

soma = lista1 + lista2

print(f"Lista 1: {lista1}")
print(f"Lista 2: {lista2}")
print()
print(f"Soma: {soma}")
print(f"Subtração: {lista1 - lista2}")
print(f"Multiplicação: {lista1 * lista2}")


divisao = lista1 / lista2
divisao_formatada = np.round(divisao, 2)

print(f"Divisão: {divisao_formatada}")
```

# 04 - Manipulação de Matrizes

Crie uma matriz 3x3 com valores aleatórios entre 0 e 1, e encontre o valor máximo e mínimo.

```python
import numpy as np

matriz = np.random.random((3, 3))
matriz_formatada = np.round(matriz, 2)

maximo = np.max(matriz_formatada)
minimo = np.min(matriz_formatada)

print(f"{matriz_formatada}")
print()
print(f"Máximo: {maximo}")
print(f"Mínimo: {minimo}")
```

# 10 - Agrupamento e Agregação

Agrupe o DataFrame pelo campo Produto e calcule a soma das quantidades vendidas para cada produto.

```python
import pandas as pd
import numpy as np

produtos = ['Notebook', 'Tablet', 'Televisão', 'Celular', 'Celular']
quantidades = np.random.randint(1, 21, size=5)
precos = np.random.randint(10, 101, size=5)

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades,
    'Preço': precos
})

df_vendas['Valor Total'] = df_vendas['Quantidade'] * df_vendas['Preço']

df_agrupado = df_vendas.groupby('Produto')['Quantidade'].sum().reset_index()

print("PRODUTOS")
print(df_vendas)
print()
print("PRODUTOS AGRUPADOS E SOMANDO A QUANTIDADE DE PRODUTOS")
print(df_agrupado)

```

# 11 - Gráfico de Linhas

Crie um gráfico de linha que mostra a variação de preços de um produto fictício ao longo de 12 meses.

```python
import pandas as pd
from matplotlib import pyplot as plt

meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']
precos = [120, 125, 130, 128, 135, 140, 138, 145, 150, 155, 160, 165]

df_precos = pd.DataFrame({
    'Mês': meses,
    'Preço': precos
})

plt.figure(figsize=(10, 6))
plt.plot(df_precos['Mês'], df_precos['Preço'], marker='o', linestyle='-', color='blue')
plt.title('Variação de Preços ao Longo de 12 Meses')
plt.xlabel('Mês')
plt.ylabel('Preço')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

# 16 - Gráfico de Barras com Seaborn

Use Seaborn para criar um gráfico de barras com os mesmos dados fictícios de quantidade vendida por produto.

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

produtos = ['Notebook', 'Tablet', 'Smartphone', 'Televisão', 'Fone de Ouvido']
quantidades = [25, 30, 15, 40, 22]

df_vendas = pd.DataFrame({
    'Produto': produtos,
    'Quantidade': quantidades
})

plt.figure(figsize=(10, 6))
sns.barplot(x='Produto', y='Quantidade', data=df_vendas)
plt.title('Quantidade Vendida por Produto')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 17 - Gráfico de Dispersão com Seaborn

Recrie o gráfico de dispersão da relação entre altura e peso usando Seaborn, incluindo uma linha de regressão.

```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

alturas = np.random.uniform(150, 200, size=10)
pesos = np.random.uniform(50, 100, size=10)

df = pd.DataFrame({
    'Altura': alturas,
    'Peso': pesos
})

plt.figure(figsize=(10, 6))
sns.regplot(x='Altura', y='Peso', data=df, scatter_kws={'color': 'blue', 'alpha': 0.7}, line_kws={'color': 'red'})
plt.title('Relação entre Altura e Peso com Linha de Regressão')
plt.grid(True)
plt.tight_layout()
plt.show()

```

# 19 - Heatmap

Crie um heatmap para mostrar a correlação entre diferentes variáveis em um DataFrame fictício.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

np.random.seed(0)
dados = pd.DataFrame({
    'Variável A': np.random.normal(size=100),
    'Variável B': np.random.normal(size=100),
    'Variável C': np.random.normal(size=100),
    'Variável D': np.random.normal(size=100)
})

correlacao = dados.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, center=0)
plt.title('Matriz de Correlação')
plt.tight_layout()
plt.show()

```

# 20 - Pairplot

Crie um pairplot para explorar relações entre várias variáveis em um conjunto de dados fictício.

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

np.random.seed(0)
dados = pd.DataFrame({
    'Variável A': np.random.normal(size=100),
    'Variável B': np.random.normal(size=100),
    'Variável C': np.random.normal(size=100),
    'Variável D': np.random.normal(size=100)
})

plt.figure(figsize=(12, 10))
sns.pairplot(dados, diag_kind='kde', plot_kws={'alpha':0.7})
plt.suptitle('Pairplot das Variáveis', y=1.02)
plt.tight_layout()
plt.show()

```
